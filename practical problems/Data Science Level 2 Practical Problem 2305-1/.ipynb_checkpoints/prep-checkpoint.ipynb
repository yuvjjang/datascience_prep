```python
colx_S = [x+'_S' for x in colx]
```

```python
df_scaled.loc[:, colx_S].sum(axis=1)
```

```python
df_scaled["INVOLVEMENT"]=df_scaled.loc[:, colx_S].sum(axis=1)
```
