```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def knn_sensor(df_train, df_test, y_param, num):
    model = KNeighborsClassifier(n_neighbors=num).fit(X=df_train.iloc[:,2:], y=df_train.loc[:,y_param])
    df_pred = pd.Series(data = model.predict(df_test.iloc[:, 2:]),\
                                               name=y_param)
    correct = (df_pred.loc[:y_param]-df_test.loc[:,y_param].reset_index(drop=True))
    correct_num = correct.value_counts().iloc[0]
    total = df_test.shape[0]
    return correct_num/total
```

```python
n_l = [2, 3, 4]
```

```python
df_TestSet5.loc[:,"LABEL"].reset_index(drop=True)
```

```python
[knn_sensor(df_train=df_TrainSet5, df_test=df_TestSet5, y_param="LABEL", num=n) for n in n_l]
```
