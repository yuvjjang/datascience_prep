```python
def standardization(X):
    m = X.mean()
    var = X.var()
    return (X-m)/(var**0.5)
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
```

```python
df_train, df_test = train_test_split(df, train_size=0.8, random_state=123)
```

```python
def kn_acc(k):
    model = KNeighborsClassifier(n_neighbors=k).fit(X=df_train.loc[:,col_x],
                                                  y=df_train["Outcome"])
    pred = model.predict(df_test.loc[:,col_x])
    acc = accuracy_score(y_true=df_test["Outcome"], y_pred=pred)
    return (k, acc)
```
