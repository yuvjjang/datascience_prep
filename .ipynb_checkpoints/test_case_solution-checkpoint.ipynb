```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("data/diabetes.csv")
```

```python
col_x=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']
```

```python
model = StandardScaler().fit(X=df.loc[:,col_x])
df.loc[:, col_x] = model.transform(X=df.loc[:, col_x])
```

```python
model = KMeans(n_clusters=2,random_state=1234).fit(X=df.loc[:,col_x])
df["label"] = model.labels_
```

```python
def weighted_sum(df, col_name):   
    vc = df.groupby(["label"])[col_name].value_counts()
    temp = pd.DataFrame(index=vc.index, columns=["counts"], data=vc.values).reset_index()
    temp["Weighted_multi"]=temp[col_name]*temp["counts"]
    temp = (temp.groupby(["label"])["Weighted_multi"].sum())/(temp.groupby(["label"])["counts"].sum())
    return pd.DataFrame(index=temp.index, columns=["Weighted_sum_avg"], data=temp.values)
```

```python
df.head()
```

```python
df_ws_glu = weighted_sum(df,"BMI")
```

```python
df_ws_bp = weighted_sum(df,"BloodPressure")
```

```python
df_ws_bmi = weighted_sum(df,"BMI")
```
