```python
dia = pd.read_csv("data/diamonds.csv")
dia.head()
```

```python
dia_price_mean = dia.groupby(["cut","color"])["price"].mean().reset_index()
dia_price_mean.head()
```

```python
dia_carat_mean = dia.groupby(["cut","color"])["carat"].mean().reset_index()
dia_carat_mean.head()
```

```python
dia_price_carat_join = pd.merge(left=dia_price_mean, right=dia_carat_mean,
                               how="inner")
```

```python
dia_price_carat_join.head()
```

```python
dia_price_carat_join["average"]= dia_price_carat_join["price"]/dia_price_carat_join["carat"]
```

```python
dia_price_carat_join.sort_values("average",ascending=False)
```
