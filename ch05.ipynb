```python
def season_len(dataframe, season):
    sum = 0
    #print(dataframe.shape)
    for num in dataframe["season"]:
        if num == season:
            sum = sum+1
    length = sum#//season
    print("number of "+"season "+str(season)+ " : "+str(length))

    return length
```

```python
df_season = df.groupby("season").sample(frac=0.05)
df_season.shape
```

```python
df_season["season"].unique()
```

```python
def each_and_total_sum(df):
    total = 0
    #print(df.shape)
    for n in df["season"].unique():
        length = season_len(df,n)
        total = total + length
    print("Total number of dataframe : "+str(total))
```
