```python
from scipy.stats import spearmanr, pearsonr
```

```python
df["is_admit"]=(df["Chance of Admit "]>0.6)+0
```

```python
df
```

```python
df["is_admit"].value_counts()
```

```python
df.groupby("is_admit")[["GRE Score", "TOEFL Score", "CGPA"]].rank(method="min")
```

```python
df_rank = df.loc[:,["GRE Score", "TOEFL Score", "CGPA"]].rank(method="min")
```

```python
dict(zip(df_rank.columns,["GRE_Rank", "TOEFL_Rank", "CGPA_Rank"]))
```

```python
df_merge = pd.concat([df, df_rank.rename(columns=dict(zip(df_rank.columns,["GRE_Rank", "TOEFL_Rank", "CGPA_Rank"])))], axis=1)
```

```python
spearmanr(df_merge["GRE_Rank"], df_merge["TOEFL_Rank"])
```

```python
spearmanr(df_merge["GRE_Rank"], df_merge["CGPA_Rank"])
```

```python
spearmanr(df_merge["CGPA_Rank"], df_merge["TOEFL_Rank"])
```
