```python
#Position_Class 결측치 처리
```

```python
def handle_mission_position_class(df):
    total_position = set(df.Position.unique())
    defender = set(pd.Series(df.Position.unique()).loc[pd.Series(df.Position.unique()).apply(lambda x: x.endswith(("B","DM")))])
    GK=set(["GK"])
    sub_position = total_position-defender-GK
    midfielder = set(pd.Series(list(sub_position)).loc[pd.Series(list(sub_position)).apply(lambda x: x.endswith("M"))])
    forward = sub_position - midfielder - GK
    df.loc[df.Position.isin(forward), "Position_Class"] = "Forward"
    df.loc[df.Position.isin(GK), "Position_Class"] = "GoalKeeper"
    df.loc[df.Position.isin(midfielder), "Position_Class"] = "Midfielder"
    df.loc[df.Position.isin(defender), "Position_Class"] = "Defender"
```

```python
#Height_cm 결측치 처리
```

```python
def handle_mission_height_cm(df):
    Height_feet_list = df.Height.loc[df.Height_cm.isna()].unique()
#     print(Height_feet_list)
    result = []
    for x in Height_feet_list:
#         print(df.Height_cm.loc[df.Height==x])
#         print(df.Height_cm.loc[df.Height==x].mode())
        result.append(df.Height_cm.loc[df.Height==x].mode().values)
#     print(result)
    df_height_missing_value = pd.DataFrame()
    df_height_missing_value["Height_cm"] = result
    df_height_missing_value["Height"] = Height_feet_list
#     display(df.loc[df.Height_cm.isna(),["Height","Height_cm"]])
    df_height_missing_value\
       .loc[df_height_missing_value["Height"]==df.Height.loc[16],"Height_cm"].values
    df_height_cm_missing = df.loc[df.Height_cm.isna(),["Height","Height_cm"]]\
    .apply(lambda x: df_height_missing_value\
           .loc[df_height_missing_value["Height"]==x["Height"],"Height_cm"].values[0], axis=1)
#     display(df.Height_cm)
#     display(df_height_cm_missing)
#     display(df_height_cm_missing.index)
#     display(df_height_cm_missing.values)
    df.loc[df_height_cm_missing.index,"Height_cm"] = df_height_cm_missing
    return df
```

```python
#Weight_lb 결측치 처리
```

```python
def handle_missing(data, ps_class, param):
    return data.loc[data.Position_Class==ps_class,param]\
    .fillna(data.loc[data.Position_Class==ps_class,param].mean(skipna=True))
```

```python
def handle_mission_weight_lb(df):
    for position_class in df.Position_Class.unique():
        df.loc[df.Position_Class==position_class,"Weight_lb"] = handle_missing(df,position_class,"Weight_lb")
    return df
```

```python
#Age 데이터 처리
```

```python
def age_mean_without_star(df, str_number):
    temp = df.Age.loc[df.Age.apply(lambda x: (x.startswith(str_number)))]
    return int(round(temp.loc[temp.apply(lambda x: not x.endswith(('*')))].astype('int32').mean(),0))
```

```python
def handle_mission_age(df):
    unique_wrong_number_list = df.Age.loc[~df.Age.apply(lambda x: x.isdecimal())].unique()
    df.Age.loc[~df.Age.apply(lambda x: x.isdecimal())]
    unique_wrong_number_mean_list = [age_mean_without_star(df,x[0]) for x in unique_wrong_number_list]
    unique_wrong_number_mean_list.sort()
    
    df.Age.loc[~df.Age.apply(lambda x: x.isdecimal())].apply(lambda x: unique_wrong_number_mean_list[int(x[0])-1])
    df.Age.loc[~df.Age.apply(lambda x: x.isdecimal())] =df.Age.loc[~df.Age.apply(lambda x: x.isdecimal())].apply(lambda x: unique_wrong_number_mean_list[int(x[0])-1])
    df.Age=df.Age.astype('int32')
```
